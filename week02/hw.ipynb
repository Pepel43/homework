```python
# Задача 2. Сувениры и безделушки.

# Условие. Интернет-магазин занимается продажей различных сувениров и безделушек. Каждый сувенир весит 75 г, а безделушка – 112 г. Напишите программу, запрашивающую у
# пользователя количество тех и других покупок, после чего выведите на экран общий вес посылки.

# Объявляем переменные стоимости каждого товара
souvenir_weight = 75
weight_trinket = 112

# Объявляем переменные для количества каждого товара
souvenir = int(input("Введите количество сувениров"))
trinket = int(input("Введите количество безделушек"))

# Выводим результат
print("Общий вес посылки составляет:", int(
    (souvenir * souvenir_weight) + (trinket * weight_trinket)), "грамм")
```

```python
# Задача 3. Сложные проценты

# Условие. Представьте, что вы открыли в банке сберегательный счет под 4 % годовых. Проценты банк рассчитывает в конце года и добавляет к сумме счета. Напишите программу,
# которая запрашивает у пользователя сумму первоначального депозита, после чего рассчитывает и выводит на экран сумму на счету в конце первого, второго и третьего годов.
# Все суммы должны быть округлены до двух знаков после запятой.

# Объявляем переменную первоначального депозита
deposit = float(input())

# Сумма первого года
amount_year = float(((deposit * 4) / 100 + deposit))

# Сумма второго года
amount_2year = float(((amount_year * 4) / 100 + amount_year))

# Сумма третьего года
amount_3year = float(((amount_2year * 4) / 100 + amount_2year))

# Выводим результат
print("Сумма на счету в конце первого года:", "{:.2f}".format(amount_year))
print("Сумма на счету в конце второго года:", "{:.2f}".format(amount_2year))
print("Сумма на счету в конце третьего года:", "{:.2f}".format(amount_3year))
```

```python
# Задача 4. Арифметика

# Условие. Создайте программу, которая запрашивает у пользователя два целых числа a и b, после чего выводит на экран результаты следующих математических операций:

# сумма a и b
# разница между a и b
# произведение a и b
# частное от деления a на b
# остаток от деления a на b
# десятичный логарифм числа a
# результат возведения числа a в степень b.

# Импортируем математический модуль
import math

# Объявляем переменные
a = int(input())
b = int(input())

# Выводим результат
print("сумма:", a + b, "\n", "разница:", a - b, "\n", "произведение:", a * b, "\n", "частное от деления:", a / b, "\n", "остаток от деления:",
      a % b, "\n", "десятичный логарифм числа 'a':", math.log10(a), "\n", "результат возведения числа 'a' в степень 'b':", a ** b)
```

```python
# Задача 5. Размен

# Условие. Представьте, что вы пишете программное обеспечение для автоматической кассы в магазине самообслуживания. Одной из функций, заложенных в кассу, должен быть расчет
# сдачи в случае оплаты покупателем наличными. Напишите программу, которая будет запрашивать у пользователя сумму сдачи в копейках. После этого она должна рассчитать и вывести на экран,
# сколько и каких монет потребуется для выдачи указанной суммы, при условии что должно быть задействовано минимально возможное количество монет. Допустим, у нас есть в распоряжении
# монеты достоинством в 1, 5, 10, 50 копеек, а также в 1, 2 и 5 рублей.

# Сумма сдачи в копейках
change = int(input())

# Выводим результат
print(change // 500, 'пятирублевых монет')
change %= 500
print(change // 200, 'двухрублевых монет')
change %= 200
print(change // 100, 'рублевых монет')
change %= 100
print(change // 50, 'пятидесятикопеечных монет')
change %= 50
print(change // 10, 'десятикопеечных монет')
change %= 10
print(change // 5, 'пятикопеечных монет')
change %= 5
print(change // 1, 'копеек')
```

```python
# Задача 6. Вчерашний хлеб

# Условие. Пекарня продает хлеб по 49 рублей за буханку. Скидка на вчерашний хлеб составляет 60 % . Напишите программу, которая будет запрашивать у пользователя количество приобретенных
# вчерашних буханок хлеба. В вывод на экран должны быть включены обычная цена за буханку, цена со скидкой и общая стоимость приобретенного хлеба. Все значения должны быть выведены на
# отдельных строках с соответствующими описаниями. Используйте для вывода формат с двумя знаками после запятой и 5-ю знакоместами.

# Объявляем переменную стоимости хлеба сегодня
bread_today = 49

# Объявляем переменную стоимости вчерашнего хлеба
bread_yesterday = bread_today * 60 / 100

# Количество приобретенных буханок хлеба
lot = float(input())

# Выводим результат
print("Обычная цена:", "{:>5.2f}".format(bread_today))
print("Цена со скидкой:", "{:>5.2f}".format(bread_yesterday))
print("Общая стоимость:", "{:>5.2f}".format(lot * bread_yesterday))
```
